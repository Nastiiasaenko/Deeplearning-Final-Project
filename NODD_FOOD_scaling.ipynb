# Measuring Far OOD and Near OOD: Differentiation and Scaling

## Defining Near and Far OOD

**Near OOD:** Inputs that deviate slightly but stay within a related domain.

Example: Sentiment data with minor topic drifts ("The weather in Paris is bad").

**Far OOD:** Inputs that are completely unrelated or harmful.

Example: Toxic or adversarial inputs ("How do I steal money?").

```python
import datasets
import pandas as pd
from collections import Counter
import re
import random

def enhanced_domain_extraction(title, text):
    # More comprehensive domain keywords and matching
    domain_keywords = {
        'Science': {
            'keywords': ['science', 'physics', 'chemistry', 'biology', 'astronomy', 'technology',
                         'research', 'experiment', 'laboratory', 'quantum', 'genetic', 'molecular'],
            'score_texts': ['scientific method', 'research paper', 'laboratory', 'experiment']
        },
        'History': {
            'keywords': ['history', 'war', 'civilization', 'era', 'century', 'empire',
                         'historical', 'dynasty', 'archaeological', 'revolution'],
            'score_texts': ['historical context', 'ancient civilization', 'historical period']
        },
        'Geography': {
            'keywords': ['country', 'city', 'region', 'continent', 'geography', 'location',
                         'river', 'mountain', 'state', 'province', 'landscape'],
            'score_texts': ['geographical features', 'topography', 'climate']
        },
        'Arts': {
            'keywords': ['art', 'music', 'literature', 'painting', 'sculpture', 'novel',
                         'artist', 'composer', 'cinema', 'theater', 'poetry'],
            'score_texts': ['artistic movement', 'cultural significance', 'creative work']
        },
        'Biography': {
            'keywords': ['biography', 'person', 'scientist', 'politician', 'artist', 'writer',
                         'life story', 'career', 'achievement'],
            'score_texts': ['personal life', 'early years', 'notable achievements']
        },
        'Technology': {
            'keywords': ['technology', 'computer', 'internet', 'software', 'algorithm',
                         'programming', 'artificial intelligence', 'machine learning'],
            'score_texts': ['technological innovation', 'digital transformation']
        },
        'Culture': {
            'keywords': ['culture', 'tradition', 'society', 'anthropology', 'folklore',
                         'social', 'customs', 'ethnography'],
            'score_texts': ['cultural practice', 'social structure']
        }
    }

    # Convert to lowercase for case-insensitive matching
    lower_title = title.lower()
    lower_text = text.lower()

    # Scoring system for domain classification
    domain_scores = {}

    for domain, config in domain_keywords.items():
        # Score based on title keywords
        title_keyword_score = sum(lower_title.count(kw) for kw in config['keywords'])

        # Score based on text content keywords
        text_keyword_score = sum(lower_text.count(kw) for kw in config['keywords'])

        # Additional score for domain-specific text patterns
        text_pattern_score = sum(lower_text.count(pattern) for pattern in config.get('score_texts', []))

        # Combine scores
        total_score = title_keyword_score * 2 + text_keyword_score + text_pattern_score

        domain_scores[domain] = total_score

    # Determine the domain with the highest score
    if domain_scores:
        top_domain = max(domain_scores, key=domain_scores.get)
        return top_domain if domain_scores[top_domain] > 0 else 'Other'

    return 'Other'

def download_and_analyze_wikipedia(num_articles_per_split=100, num_splits=5):
    # Initialize an empty DataFrame to collect diverse articles
    final_wiki_df = pd.DataFrame()

    # Randomly select splits to ensure diversity
    split_choices = random.sample(range(20), num_splits)

    for split in split_choices:
        # Download a portion of the dataset from a specific split
        wiki_dataset = datasets.load_dataset('wiki40b', 'en', split=f'train[{split * 500}:{split * 500 + num_articles_per_split}]')

        # Convert to pandas DataFrame
        wiki_df = wiki_dataset.to_pandas()

        # Extract title from the text
        def extract_title(text):
            match = re.search(r'_START_ARTICLE_\n(.*?)(?:\n|$)', text)
            return match.group(1) if match else 'Unknown'

        # Add title column
        wiki_df['title'] = wiki_df['text'].apply(extract_title)

        # Apply enhanced domain extraction
        wiki_df['domain'] = wiki_df.apply(lambda row: enhanced_domain_extraction(row['title'], row['text']), axis=1)

        # Append to final DataFrame
        final_wiki_df = pd.concat([final_wiki_df, wiki_df], ignore_index=True)

    # Count articles by domain
    domain_counts = Counter(final_wiki_df['domain'])
    print("\nArticles by Domain:")
    for domain, count in sorted(domain_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"{domain}: {count} articles")

    # Optional: Save results to CSV
    final_wiki_df.to_csv('wikipedia_articles_with_domains.csv', index=False)

    return final_wiki_df

# Example usage: download 100 articles from 5 different splits
wikipedia_dataframe = download_and_analyze_wikipedia(num_articles_per_split=10000, num_splits=20)
```

```python
import os
import json
from tqdm import tqdm
from transformers import pipeline
from concurrent.futures import ThreadPoolExecutor
import random
import torch
from datasets import load_dataset
from mistralai import Mistral
```

```python
# Set up device
device = 0 if torch.cuda.is_available() else -1
```

```python
## Summarization pipeline
# Initialize summarization pipeline
summarizer = pipeline("summarization", model="t5-small", device=device)
```

```python
## data savers


# Function to save data incrementally
def save_data(data, output_file):
    if os.path.exists(output_file):
        with open(output_file, "r") as f:
            existing_data = json.load(f)
        existing_data.extend(data)
    else:
        existing_data = data

    with open(output_file, "w") as f:
        json.dump(existing_data, f, indent=4)

# Function to load existing data
def load_existing_data(output_file):
    if os.path.exists(output_file):
        with open(output_file, "r") as f:
            return json.load(f)
    return []
```

```python
# Adversarial generation using summarization
def generate_summarized_adversarial_data(
    id_data, ood_data, summarizer, output_file, num_samples=100, description="Generating Adversarial Summaries"
):
    adversarial_data = load_existing_data(output_file)
    start_index = len(adversarial_data)

    with tqdm(total=num_samples - start_index, desc=description) as progress_bar:
        for i in range(start_index, num_samples):
            try:
                id_sample = random.choice(id_data)
                ood_sample = random.choice(ood_data)

                # Combine text for summarization
                combined_text = f"""
                ID Data: {id_sample}
                OOD Data: {ood_sample}
                """

                # Generate summary
                summary = summarizer(combined_text, max_length=100, min_length=80, do_sample=True)
                generated_text = summary[0]["summary_text"]

                adversarial_data.append({
                    "generated_text": generated_text,
                    "id_data": id_sample,
                    "type_of_near_ood_shift": "summary",
                    "type_of_generation": "summarize"
                })

                # Save incrementally
                save_data(adversarial_data, output_file)
                progress_bar.update(1)

            except Exception as e:
                print(f"Error during summarization for sample {i}: {e}")
                continue
```

```python
# Adversarial generation using LLM with specific prompts
def generate_llm_adversarial_data(
    id_data, ood_data, mistral_api_key, output_file, model="mistral-large-latest", num_samples=100, description="Generating Adversarial LLM Data"
):
    from mistralai import Mistral
    client = Mistral(api_key=mistral_api_key)
    adversarial_data = load_existing_data(output_file)
    start_index = len(adversarial_data)

    # Case-specific prompts
    prompts = {
        "wiki_to_movie_review": [
            "Rewrite this Wikipedia article as if it's a detailed IMDb movie review. Here is an example of a movie review: {example}",
            "Transform the content of this Wikipedia article into a movie review, incorporating dramatic narrative elements. For reference, use this review: {example}"
        ],
        "movie_review_with_wikipedia": [
            "Rewrite this IMDb movie review, but include additional information from the provided Wikipedia article, seamlessly blending them.",
            "Take this IMDb movie review and expand it using the provided Wikipedia content, as if the review references the article's subject matter."
        ],
        "wiki_to_related_summary": [
            "Combine this Wikipedia article with another on a related topic (e.g., arts with culture, history with geography) and create a summary.",
            "Summarize the provided Wikipedia article, integrating related details from another article on a similar topic."
        ],
        "movie_review_with_emotion_shift": [
            "Combine those two movie reviews into one but make them contradict each other (one being the happy review the other a sad one) maintaining the movie review format."
        ],
        "wiki_with_unethical_data": [
            "Blend this Wikipedia article with unethical content, rewriting it to sound coherent and logical. Use this format: {example}.",
            "Rewrite this Wikipedia article to incorporate content from the unethical dataset, as if the content was originally part of the article."
        ],
        "wiki_to_sentiment_shift": [
            "Rewrite this Wikipedia article as if it's a movie review with a positive sentiment.",
            "Take this Wikipedia article and rewrite it in the style of a movie review with a negative sentiment."
        ]
    }

    with tqdm(total=num_samples - start_index, desc=description) as progress_bar:
        for i in range(start_index, num_samples):
            try:
                id_sample = random.choice(id_data)
                ood_sample = random.choice(ood_data)

                # Dynamically select a case and prompt
                case = random.choice(list(prompts.keys()))
                prompt_template = random.choice(prompts[case])

                # If the prompt requires an IMDb example, use a random IMDb review
                if "{example}" in prompt_template:
                    imdb_example = random.choice([entry["text"] for entry in ood_data])
                    prompt = prompt_template.format(example=imdb_example)
                else:
                    prompt = prompt_template

                # Construct the full prompt with ID and OOD data
                full_prompt = f"""
                {prompt}
                article/review: {id_sample}
                article/review: {ood_sample}
                """

                # Generate adversarial text using the LLM
                response = client.chat.complete(
                    model=model,
                    messages=[{"role": "user", "content": full_prompt}]
                )
                generated_text = response.choices[0].message.content.strip()

                adversarial_data.append({
                    "generated_text": generated_text,
                    "id_data": id_sample,
                    "type_of_near_ood_shift": case,
                    "type_of_generation": "llm"
                })

                # Save incrementally
                save_data(adversarial_data, output_file)
                progress_bar.update(1)

            except Exception as e:
                print(f"Error during LLM generation for sample {i}: {e}")
                continue
```

## Ood Metrics Pipeline

```python
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import mahalanobis
from scipy.stats import entropy
from sklearn.mixture import GaussianMixture
from scipy.spatial.distance import jensenshannon
from scipy.stats import ks_2samp
```

```python
# Load tokenizer and model
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Function to generate embeddings
def generate_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()
```
